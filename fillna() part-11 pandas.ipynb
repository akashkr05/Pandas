# 2. method('ffill', 'pad', 'bfill', 'backfill', 'None')

```python
df.fillna(method = 'ffill')
```

```python
df
```

```python
df.fillna(method = 'bfill')
```

```python
df.fillna(method = 'backfill')
```

```python
df.fillna(method = 'pad')
```

# 3. axis = 0 for row and 1 for column

```python
df.fillna(0, axis = 0)
```

```python
df.fillna(0, axis = 1)
```

```python
df.fillna(method = 'ffill', axis = 0)
```

# 4.limit = int(always take more than 0)

```python
df
```

```python
df.fillna(0, limit = 1)
```

```python
df.fillna(0, limit = 2)
```

```python
df.fillna(method = 'ffill', limit = 1)
```

# inplace = boolean(True/False)

```python
df
```

```python
df.fillna(0, inplace = True)
df
```

```python
df.fillna(0, inplace = False)
```
